```python
train_path="D://ML/datasets/Food-5K/train"
test_path="D://ML/datasets/Food-5K/test"
```

```python
IMAGE_SIZE=[200,200]
```

```python
#useful for getting number of files
image_files=glob(train_path+'/*/*.jpg')
valid_files=glob(test_path+'/*/*.jpg')
```

```python
len(image_files), len(valid_files)
```

```python
#useful to get number of classes
folders=glob(train_path+'/*')
folders
```

```python
plt.imshow(image.load_img(np.random.choice(image_files)))
plt.show()
```

```python
ptm=PretrainedModel(
    input_shape=IMAGE_SIZE+[3],
    weights='imagenet',
    include_top=False)
```

```python
ptm.trainable=False
```

```python
K=len(folders)

x=Flatten()(ptm.output)
x=Dense(K, activation='softmax')(x)
```

```python
model=Model(inputs=ptm.input, outputs=x)
```

```python
model.summary()
```

```python
gen=ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    preprocessing_function=preprocess_input
)
```

```python
batch_size=128

train_generator=gen.flow_from_directory(
    train_path,
    shuffle=True,
    target_size=IMAGE_SIZE,
    batch_size=batch_size,
)
valid_generator=gen.flow_from_directory(
    test_path,
    target_size=IMAGE_SIZE,
    batch_size=batch_size,
)
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
r=model.fit_generator(
    train_generator,
    validation_data=valid_generator,
    epochs=5,
    steps_per_epoch=int(np.ceil(len(image_files)/batch_size)),
    validation_steps=int(np.ceil(len(valid_files)/batch_size)),
)
```

```python
plt.plot(r.history['loss'], label='loss')
plt.plot(r.history['val_loss'], label='val_loss')
plt.show()
```
