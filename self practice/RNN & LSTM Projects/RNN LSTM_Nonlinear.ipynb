```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
from tensorflow.keras.layers import Input, SimpleRNN, Dense, Flatten, LSTM
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD, Adam
```

## Making some synthetic data

```python
series=np.sin((0.1*np.arange(400))**2)
```

```python
plt.plot(series)
plt.show()
```

```python
T=10
D=1
X=[]
Y=[]
for t in range(len(series)-T):
    x=series[t:t+T]
    X.append(x)
    y=series[t+T]
    Y.append(y)
    
X=np.array(X).reshape(-1,T)
Y=np.array(Y)
N=len(X)

print(X.shape, Y.shape)
#X is only N*T as we are using linear regression and it accepts data in the form of N*D rectangle and not N*T*D
```

```python
i = Input(shape=(T,))
x = Dense(1)(i)
model=Model(i,x)
model.compile(loss='mse',
             optimizer=Adam(lr=0.01))
```

```python
r= model.fit(X[:-N//2],Y[:-N//2], epochs=80, validation_data=(X[-N//2:], Y[-N//2:]))
```

```python
plt.plot(r.history['loss'],label='loss')
plt.plot(r.history['val_loss'],label='val_loss')
plt.legend()
```

## One-step forecast using true targets

```python
output=model.predict(X)
print(output.shape)
predictions=output[:,0]

plt.plot(Y, label='targets')
plt.plot(predictions, label='predictions')
plt.title("Linear Regression Predictions")
plt.legend()
plt.show()
```

## One-step forecast using another method

```python
validation_target=Y[-N//2:]
validation_predictions=[]
i=-N//2
while(len(validation_predictions)<len(validation_target)):
    pred=model.predict(X[i].reshape(1,-1))[0,0]
    validation_predictions.append(pred)
    i+=1
plt.plot(validation_target,label='validaion_target')
plt.plot(validation_predictions,label='validation_predictions')
plt.legend()
plt.show()
```

## Multi step forecast (correct method)

```python
validation_target=Y[-N//2:]
validation_predictions=[]

last_x=X[-N//2]
while(len(validation_predictions)<len(validation_target)):
    pred=model.predict(last_x.reshape(1,-1))[0,0]
    validation_predictions.append(pred)
    
    last_x=np.roll(last_x,-1)
    last_x[-1]=pred
plt.plot(validation_target,label='validation_target')
plt.plot(validation_predictions,label='validation_predictions')
plt.legend()
plt.show()
```

# RNN

```python
X=X.reshape(-1,T,1) #make dataset ok for RNN i.e. N*T*D
```

```python
i=Input(shape=(T,D))
x=SimpleRNN(10)(i)
x=Dense(1)(x)
model=Model(i,x)

model.compile(loss='mse',
             optimizer=Adam(lr=0.05))
```

```python
r=model.fit(X[:-N//2],Y[:-N//2], epochs=200, batch_size=32, validation_data=(X[-N//2:], Y[-N//2:]))
```

```python
plt.plot(r.history['loss'],label='loss')
plt.plot(r.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```

## One step forecast

```python
output=model.predict(X)
#print(output.shape)
pred=output[:,0]

plt.plot(Y, label='target')
plt.plot(pred, label='pred')
plt.legend()
plt.show()
```

## Multi step forecast

```python
validation_target=Y[-N//2:]
validation_predictions=[]

last_x=X[-N//2]
while(len(validation_predictions)<len(validation_target)):
    pred=model.predict(last_x.reshape(1,-1,1))[0,0]
    validation_predictions.append(pred)
    
    last_x=np.roll(last_x,-1)
    last_x[-1]=pred
plt.plot(validation_target,label='validation_target')
plt.plot(validation_predictions,label='validation_predictions')
plt.legend()
plt.show()
```
